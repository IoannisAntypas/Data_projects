# "Fear and joy in Jane Austen vs H.P. Lovecraft"



```python
%matplotlib inline
import pandas as pd
import glob
from sklearn.feature_extraction.text import TfidfVectorizer

```

```python
filenames = glob.glob("lovecraft/*.txt")
contents = [open(file).read() for file in filenames]
lovecraft_df = pd.DataFrame({
    'filename': filenames,
    'body': contents,
    'author': 'Lovecraft'
})
lovecraft_df.head()
```

```python
filenames = glob.glob("austen/*.txt")
contents = [open(file).read() for file in filenames]
austen_df = pd.DataFrame({'filename':filenames, 'body': contents, 'author': 'Austen'})
austen_df.head()
```

```python
# df = pd.concat([lovecraft_df, austen_df], ignore_index=True)
```

```python
filepath = "NRC-Emotion-Lexicon-v0.92/NRC-emotion-lexicon-wordlevel-alphabetized-v0.92.txt"
emolex_df = pd.read_csv(filepath,  names=["word", "emotion", "association"], skiprows=45, sep='\t')
emolex_df = emolex_df.pivot(index='word', columns='emotion', values='association').reset_index()
emolex_df.head()
```

```python
scary_words = emolex_df[emolex_df.fear == 1]['word']
happy_words = emolex_df[emolex_df.joy == 1]['word']


```

```python
vec = TfidfVectorizer(vocabulary=emolex_df.word,
                      use_idf=False, 
                      norm='l1') # ELL - ONE
matrix = vec.fit_transform(austen_df['body'])
vocab = vec.get_feature_names()
wordcount_df = pd.DataFrame(matrix.toarray(), columns=vocab)
wordcount_df.head()
```

```python
lovecraft_df['joy'] = wordcount_df[happy_words].sum(axis=1)
lovecraft_df['fear'] = wordcount_df[scary_words].sum(axis=1)
lovecraft_df.head()

```

```python
vec = TfidfVectorizer(vocabulary=emolex_df.word,
                      use_idf=False, 
                      norm='l1') # ELL - ONE
matrix = vec.fit_transform(lovecraft_df['body'])
vocab = vec.get_feature_names()
wordcount_df = pd.DataFrame(matrix.toarray(), columns=vocab)
wordcount_df.head()
```

```python
austen_df['joy'] = wordcount_df[happy_words].sum(axis=1)
austen_df['fear'] = wordcount_df[scary_words].sum(axis=1)
austen_df
```

```python
ax = austen_df.plot(x='joy', y='fear', kind='scatter')
lovecraft_df.plot(x='joy', y='fear', kind='scatter',c='red', ax=ax)
```
